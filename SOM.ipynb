```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import somoclu
%matplotlib inline
```

```python
c1 = np.genfromtxt('Blood Transfusion Service Center.csv', delimiter=',', usecols=(0, 1, 2, 3, 4), max_rows= 50) 
c2 = np.genfromtxt('Blood Transfusion Service Center.csv', delimiter=',', usecols=(0, 1, 2, 3, 4),skip_header= 50, max_rows = 50)
c3 = np.genfromtxt('Blood Transfusion Service Center.csv', delimiter=',', usecols=(0, 1, 2, 3, 4),skip_header=100, max_rows = 50)
c4 = np.genfromtxt('Blood Transfusion Service Center.csv', delimiter=',', usecols=(0, 1, 2, 3, 4),skip_header=150, max_rows = 50)
data = np.float32(np.concatenate((c1, c2, c3,c4)))
```

```python
colors = ["red"] * 50
colors.extend(["green"] * 50)
colors.extend(["blue"] * 50)
colors.extend(["yellow"] * 50)
```

```python
fig = plt.figure()
ax = Axes3D(fig)
ax.scatter(data[:, 0], data[:, 1], data[:, 2], data[:, 3], c=colors)
labels = range(800)
```

```python
n_rows, n_columns = 100, 160
som = somoclu.Somoclu(n_columns, n_rows, data=data)
%time som.train()
```

```python
som.view_component_planes()
```

```python
som.view_umatrix(bestmatches=True, bestmatchcolors=colors, labels=labels)
```
