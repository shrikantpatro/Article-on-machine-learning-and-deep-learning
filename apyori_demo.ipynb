```python
# Import the libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# Import Dataset 
dataset = pd.read_excel(r"C:\Users\shrikant\Desktop\content\SEM2\Machine Learning\LAB\Dataset\Online_Retail.xlsx", header = None)
records = []
for i in range(0, 541909):
    records.append([str(dataset.values[i,j]) for j in range(0, 8)])
```

```python
# Train Apriori Model
from apyori import apriori
rules = apriori(records, min_support = 0.003, min_confidence = 0.2, min_lift = 3, min_length = 2)

```

```python
list(rules)
```

```python
# Visualising the results
results = list(rules)
```
