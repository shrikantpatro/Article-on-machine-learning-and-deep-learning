```python
import pandas as pd
import numpy as np
from sklearn import svm, datasets
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data=pd.read_csv('/home/shrikant/Downloads/Blood Transfusion Service Center.csv')
```

```python
data.head()
```

```python
X=data.iloc[:,1:3]
X1=data.iloc[:,2:4]
y=data.iloc[:,-1]
X
```

```python
x_min, x_max = data.iloc[:, 1].min(), data.iloc[:, 1].max() + 1
y_min, y_max = data.iloc[:, 2].min(), data.iloc[:, 2].max() + 1
h = (x_max / x_min)/100
xx, yy = np.meshgrid(np.arange(x_min, x_max, h),
np.arange(y_min, y_max, h))
X_plot = np.c_[xx.ravel(), yy.ravel()]
```

```python
# Create the SVC model object
C = 1.0 # SVM regularization parameter
svc = svm.SVC(kernel='linear', C=C, decision_function_shape='ovr').fit(X, y)
Z = svc.predict(X_plot)
Z = Z.reshape(xx.shape)
```

```python
plt.figure(figsize=(15, 5))
plt.subplot(121)
plt.contourf(xx, yy, Z, cmap=plt.cm.tab10, alpha=0.3)
plt.scatter(data.iloc[:, 1], data.iloc[:, 2], c=y, cmap=plt.cm.Set1)
plt.xlabel('Frequency (times)')
plt.ylabel('Monetary (c.c. blood)')
plt.xlim(xx.min(), xx.max())
plt.title('SVC with linear kernel')
```

```python
# Create the SVC model object
C = 1.0 # SVM regularization parameter
svc = svm.SVC(kernel='rbf', C=C, decision_function_shape='ovr').fit(X, y)
```

```python
Z = svc.predict(X_plot)
Z = Z.reshape(xx.shape)
```

```python
plt.subplot(122)
plt.contourf(xx, yy, Z, cmap=plt.cm.tab10, alpha=0.3)
plt.scatter(data.iloc[:, 1], data.iloc[:, 2], c=y, cmap=plt.cm.Set1)
plt.xlabel('Monetary (c.c. blood)')
plt.ylabel('Time (months)')
plt.xlim(xx.min(), xx.max())
plt.title('SVC with RBF kernel')
plt.show()
```
